# 不稳定供货商的最大供货量预测--SARIMA

```python
unstable_list = ['S140', 'S282', 'S108', 'S308', 'S330', 'S284', 'S365', 'S040', 'S364', 'S055', 'S114', 'S007', 'S086', 'S291', 'S338', 'S150', 'S129']
stable_list = [i for i in top_50_ids if i not in unstable_list]
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX

def predict_supplier_supply_next_24_weeks(supplier_id, supply_df, plot=True):
    row = supply_df[supply_df.iloc[:, 0] == supplier_id]
    if row.empty:
        raise ValueError(f"未在供货量表中找到供应商 {supplier_id}")
    supply_series = row.iloc[0, 2:].astype(float).values
    weeks = pd.date_range(start='2017-01-01', periods=len(supply_series), freq='W')
    supply_ts = pd.Series(supply_series, index=weeks)
    order = (1, 1, 1) 
    seasonal_order = (1, 1, 1, 48)

    with warnings.catch_warnings():
        warnings.simplefilter("ignore")
        model = SARIMAX(supply_ts, order=order, seasonal_order=seasonal_order)
        results = model.fit(disp=False)

    forecast = results.get_forecast(steps=24)
    predicted_supply = forecast.predicted_mean.values
    predicted_supply = np.where(predicted_supply < 0, 0, predicted_supply).astype(int)
    if plot:
        plt.figure(figsize=(12, 6))
        plt.plot(supply_ts[-100:], label='历史供货量', marker='o', markersize=3)
        future_index = pd.date_range(start=supply_ts.index[-1], periods=25, freq='W')[1:]
        plt.plot(future_index, forecast.predicted_mean, label='预测供货量', color='r', linestyle='--', marker='x', markersize=4)
        pred_ci = forecast.conf_int()
        plt.fill_between(future_index,
                         pred_ci.iloc[:, 0], pred_ci.iloc[:, 1],
                         color='pink', alpha=0.3, label='95% 置信区间')
        plt.title(f'供应商 {supplier_id} 的供货量预测')
        plt.xlabel('时间')
        plt.ylabel('供货量 ($m^3$)')
        plt.legend()
        plt.grid(True)
        plt.tight_layout()
        plt.show()
    return predicted_supply
```

```python
show = True
for idx, supplier_id in enumerate(unstable_list):
    if idx >= 2:
        show = False
    unstable_forecast_supply = predict_supplier_supply_next_24_weeks(supplier_id, supply_df, plot=show)
    future_index = [f'W{i+241}' for i in range(24)]
    forecast_supply_top50.loc[forecast_supply_top50["供应商ID"] == supplier_id, future_index] = unstable_forecast_supply
    # print(f"{supplier_id}未来 24周供货量预测结果：", unstable_forecast_supply)
```

# 稳定供货商的最大供货量预测--采用最大供货量

```python
future_index = [f'W{i+241}' for i in range(24)]
for supplier_id in stable_list:
    row = supply_df[supply_df.iloc[:, 0] == supplier_id]
    if not row.empty:
        max_supply = row.iloc[0, 2:].max()
        forecast_supply_top50.loc[forecast_supply_top50["供应商ID"] == supplier_id, future_index] = [max_supply] * 24
        print(f"{supplier_id} 的最大供货量：{max_supply}")
    else:
        print("ERROR! with empty!")
```

# SARIMA订货量

```python
week_columns = [f'W{i:03d}' for i in range(1, 241)]
abc_order_data = order_df.groupby('材料分类')[week_columns].sum().loc[['A', 'B', 'C']]
weeks = pd.date_range(start='2017-01-01', periods=240, freq='W')
abc_order_ts = pd.DataFrame({
    'A': abc_order_data.loc['A'].values,
    'B': abc_order_data.loc['B'].values,
    'C': abc_order_data.loc['C'].values
}, index=weeks)
abc_order_ts["sum"] = abc_order_ts["A"] * 1.2 + abc_order_ts["B"] * (11 / 12) + abc_order_ts["C"]
abc_order_ts
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
warnings.filterwarnings("ignore")

def predict_weekly_order_for_category(ts_series, steps=24, order=(1, 1, 1), seasonal_order=(2, 1, 1, 48)):
    model = SARIMAX(ts_series, order=order, seasonal_order=seasonal_order)
    result = model.fit(disp=False)
    forecast = result.get_forecast(steps=steps)
    predicted_mean = forecast.predicted_mean.values
    predicted_mean = np.where(predicted_mean < 0, 0, predicted_mean)  # 替换负值为0
    return forecast.predicted_mean, forecast.conf_int()

future_weeks = pd.date_range(start=weeks[-1] + pd.Timedelta(days=7), periods=24, freq='W')

mean, ci = predict_weekly_order_for_category(abc_order_ts["sum"], steps=24)
predicted_means = mean.values
conf_ints= ci.values
```

```python
order_predicted_df = pd.DataFrame({'预测订货量': predicted_means}, index=future_weeks)
new_index = [f"W{i}" for i in range(241, 265)]
order_predicted_df["Week"] = new_index
print(order_predicted_df)
order_predicted_df.to_excel("预测订货总量结果.xlsx", index=False)
```

```python
plt.rcParams.update({
    'font.size': 14,
    'lines.linewidth': 2,
    'lines.markersize': 6,
    'xtick.direction': 'in',
    'ytick.direction': 'in',
    'xtick.top': True,
    'ytick.right': True,
    'axes.spines.top': True,
    'axes.spines.right': True,
    'legend.frameon': False,
    'axes.grid': True,
    'grid.linestyle': '--',
    'grid.alpha': 0.7
})
import matplotlib.pyplot as plt

plt.figure(figsize=(18, 7))  # 调整大小

plt.plot(abc_order_ts["sum"][-100:], label='历史订货量', color='blue', marker='o', markersize=3)
plt.plot(order_predicted_df["预测订货量"], 
         label='预测订货量', color='red', linestyle='--', marker='x', markersize=5)
plt.title('未来24周订货总量预测')
plt.ylabel('订货量 ($m^3$)')
plt.xlabel('时间')
plt.legend()
plt.grid(True)
plt.xticks([])
plt.tight_layout()
plt.show()
```

# 模拟求解

```python
predict_order = pd.read_excel("预测订货总量结果.xlsx")
predict_supply = pd.read_excel("供应商供货量预测结果用C代替.xlsx")
predict_order_capacity = predict_order
predict_order_capacity["预测订货量"] = predict_order_capacity["预测订货量"] / 0.72
predict_supply_capacity_top50 = predict_supply
predict_supply_capacity_top50.iloc[:, 1:-1] = predict_supply.iloc[:, 1:-1] / 0.72
```

```python
num_suppliers = 16
break_flag = True
while break_flag:
    random_suppliers = predict_supply_capacity_top50["供应商ID"].head(num_suppliers)
    sum = 0
    for i in random_suppliers:
        sum += predict_supply_capacity_top50.loc[predict_supply_capacity_top50["供应商ID"] == i]["W241"].values
        if sum > 28200:
            print(f'{random_suppliers.to_list()}, sum capacity = {sum}')
            select_suppliers_df = predict_supply_capacity_top50[predict_supply_capacity_top50['供应商ID'].isin(random_suppliers)]
            break_flag = False
            break
```

```python

class Status:
    def __init__(self, suppliers_list, week):
        self.week = week
        self.suppliers_list = suppliers_list
        self.inventory_capacity = 0         # 库存产能
        self.order_capacity = 0             # 订单产能
        self.received_capacity = 0          # 到货产能
        self.transform = []                 # 转运方案
        self.max_supply_capacity = select_suppliers_df[self.week].astype(float).sum()        # 最大供货能力
    
    def update_inventory_capacity(self):
        self.inventory_capacity += self.received_capacity - 28200
```

```python

%matplotlib inline
from IPython.display import clear_output, display
import time
import random

initial_inventory = 56400
current_inventory = initial_inventory
history_maxSupply = []
history_inventory = []
history_received = []


fig, ax = plt.subplots(figsize=(10, 6))
line_max_supply, = ax.plot([], [], label='最大供货总产能', color='blue')
line_inventory, = ax.plot([], [], label='库存产能', color='gray', linestyle='--')
line_received, = ax.plot([], [], label='收货产能', color='orange')


ax.set_xlabel("时间")
ax.set_ylabel("产能 ($m^3$)")
ax.set_title("库存与产能仿真")
ax.grid(False)
ax.legend()
ax.set_ylim(0, initial_inventory * 1.2)
display(fig)

for week_idx in range(24):
    week = "W" + str(week_idx + 241)
    s = Status(random_suppliers, week)
    s.inventory_capacity = current_inventory
    
    history_inventory.append(s.inventory_capacity)
    
    s.received_capacity = s.max_supply_capacity * random.uniform(0.9, 1)
    
    # 更新库存
    s.update_inventory_capacity()
    current_inventory = s.inventory_capacity
    
    
    # 保存数据
    history_maxSupply.append(s.max_supply_capacity)
    history_received.append(s.received_capacity)

    ax.axhline(y=56400, color='red', linestyle='dotted', label='初始库存')
    ax.axhline(y=28200, color='green', linestyle='dotted', label='两周产能下限')
    
    line_max_supply.set_xdata(range(week_idx+1))
    line_max_supply.set_ydata(history_maxSupply)
    line_inventory.set_xdata(range(week_idx+1))
    line_inventory.set_ydata(history_inventory)
    line_received.set_xdata(range(week_idx+1))
    line_received.set_ydata(history_received)
    
    weeks_labels = [f"W{i+241}" for i in range(week_idx+1)]
    ax.set_xticks(range(week_idx+1))
    ax.set_xticklabels(weeks_labels, rotation=45)
    ax.set_xlim(0, 23)
    ax.set_ylim(0, 90000)
    ax.set_title(f"第{week_idx + 1}周 Status")
    clear_output(wait=True)
    display(fig)
    time.sleep(0.1)
plt.close()
```
